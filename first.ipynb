## Guide to Jupter notebooks

Do `Jupyter notebook` to create a new notebook, which will launch a note book server. Don't close this server! 

Each notebook needs a kernel e.g. python to run the code snippets

It is recommended that you use Anaconda to have all packages installed, and run code snippets using (base) conda, as setting up everything in powershell can be a bit a pain.

If you want to have a lightweight work environment then you might just want to have condalight instead. [Guide to installing condamini](https://educe-ubc.github.io/conda.html)

```python
# You could do leetcode questions in a jupyter notebook to store notes.

def get_smallest(ray : str) -> str:
    used = [False for s in ray]
    stack = []
    result = []
    top = -1
    current_lex = 0
    

    # invariant: Pushing char onto stack means that it will be inserted before
    # all characters before it. Greedy solution -> Ensure char > all chars before
    # Cannot have possible solution that is more optimal.
    # i.e. push onto stack if we see a char that is larger than top then pop the top
    # and append it to kevin, and keep doing it until we get to char with val less than
    # -> like the airplane tower reachability question.
 
    for i in range(len(ray)):
        char = ray[i]
        while (ord(char) > top and len(stack) != 0):
            to_append = stack.pop();
            if len(stack) != 0:
                new_top = stack.pop()
                top = ord(new_top)
                stack.append(new_top);
            else:
                top = -1
            result.append(to_append)
        stack.append(char)
        top = ord(char)
        # print(stack)
    
    while len(stack) != 0:
        result.append(stack.pop())
    
    return "".join(result)

def main():
    result = get_smallest("gets")
    print(f"The result is {result}")

main()
    
```
